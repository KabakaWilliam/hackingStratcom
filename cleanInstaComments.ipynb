```python
with open("billMelindaIG.txt", "r") as f:
    comments = f.readlines()
```

```python
def comment_cleaner(comments):
    store = []
    for comment in comments:
        store.append(comment.replace('\n', ''))
    return store
```

```python
# for comments in cleanedComments:
#     if len(comments) > 0:
#         pass
#     else: 
#         cleanedComments.remove(comments)

def removeEmptyComments(comments):
    for comment in comments:
        if len(comment) > 0:
            pass
        elif len(comment) == 0:
            comments.remove(comment)
        else: 
            comments.remove(comment)
    return comments

```

```python
# will remove all the excess info(name and dates) and return only comment
def refineComments(comments):
    comments = comments[2::2]
    return comments
```

```python
def cleanInstaComments(fileName):
    store = []
    with open(fileName, "r") as f:
        comments = f.readlines() 
    FirstFilteredComments = comment_cleaner(comments)
    SeconfFilteredComments = removeEmptyComments(FirstFilteredComments)
    AlmostComments = refineComments(SeconfFilteredComments)
    for comments in AlmostComments:
        if "profile picture" not in comments:
            if "wReply" not in comments:
                store.append(comments)
    CleanedComments = "".join(store)
    with open("Cleaned" + fileName , "w") as f:
        for comment in CleanedComments:
            f.write(comment)
```
